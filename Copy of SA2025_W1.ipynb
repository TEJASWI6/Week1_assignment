### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
best_cars = (
    df[df['mpg']>df['mpg'].median()]
    .assign(hp_per_weight = lambda x:x['horsepower']/x['weight'])
)

max_ratio= best_cars['hp_per_weight'].max()

# Since 'name' is index, get index values where ratio == max
best_car_names = best_cars[best_cars['hp_per_weight'] == max_ratio].index.values

print("Car(s) with best horsepower-to-weight ratio and above median mpg:")
print(best_car_names)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import matplotlib.pyplot as plt
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().unstack()

avg_mpg.plot(figsize=(10,6), marker='o')

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()

```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
import seaborn as sns
import matplotlib.pyplot as plt

df['origin'] = df['origin'].str.lower()

plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    sizes=(20, 200),
    hue_order=['japan', 'europe', 'usa'],
    palette='deep',
    alpha=0.7
)

plt.title('Scatterplot of Horsepower vs Weight\nColored by Origin and Sized by MPG')
plt.xlabel('Horsepower', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.legend(title='Origin')

plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
print(type(df))
```

```python
# Assuming:
# - 'name' is set as the DataFrame index


# Step 1: Count unique years each car appears in
name_year_counts = df.groupby(df.index)['model_year'].nunique()

# Step 2: Select cars that appear in 2 or more years
multi_year_names = name_year_counts[name_year_counts >= 2].index

# Step 3: Filter DataFrame to only those cars
multi_year_df = df.loc[multi_year_names]

# Step 4: Calculate std deviation of mpg for each car model
mpg_std = multi_year_df.groupby(multi_year_df.index)['mpg'].std()

# Step 5: Filter cars with std deviation less than 1.0 (low variation)
low_variation_names = mpg_std[mpg_std < 1.0].index

# Step 6: Filter DataFrame again for low variation cars
consistent_cars_df = multi_year_df.loc[low_variation_names]

result = consistent_cars_df.groupby(consistent_cars_df.index).agg(
    appearances=('model_year', 'count'),
    avg_mpg=('mpg', 'mean'),
    mpg_std=('mpg', 'std')
).reset_index().rename(columns={'index': 'name'})

# Step 8: Sort results by appearances (desc), then avg_mpg (desc)
result_sorted = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

print(result_sorted)
```
